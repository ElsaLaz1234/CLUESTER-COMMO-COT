# Do COT Positioning Clusters Anticipate Commodity Return Co-movement?
### Results & Interpretation
---
**Universe:** 9 commodity futures (CL, NG, RB, GC, SI, HG, ZC, ZW, ZS) · 2006–2024 · Weekly
**Method:** Ward hierarchical clustering (k=3, correlation distance) · 12-month rolling window
**Signal:** ARI between COT clusters at t−h and return clusters at t · lags h ∈ {0, 1m, 2m, 3m}


---
# Part I — Qualitative Evidence: Dendrogram Snapshots

---
## Crisis Periods — Peak Signal Weeks
### Peak Crisis — ARI 0.357 to 0.739

Oil crash 2014 and Gas shock 2021 show the strongest signal. During gradual commodity cycles,
managed money build complex cross-sectoral views — grouping GC with cereals, SI with industrial
metals — that transcend natural sector boundaries. These cross-sectoral COT clusters at t−2m are
confirmed by return co-movement at t with high precision (ARI up to 0.739). The signal is strong
precisely because the repositioning is progressive, informational, and sustained over several months.

COVID 2020 and GFC 2008 peaks show weaker signal (ARI 0.357 and 0.416). The COT structure
at t−2m reflects positioning *before* the shock — managed money had no information about the
impending crash. Return clusters post-shock revert to natural sectoral structure (energy down
together, safe havens up together) driven by a single risk-off factor, which diverges from the
pre-shock COT positioning.

> **Key pattern:** cross-sectoral COT groupings predict cross-sectoral return co-movement during
> gradual cycles. During acute crashes, COT positioning is pre-shock and non-informational.

## Crisis Periods — Trough Signal Weeks
### Trough Crisis — ARI −0.246 to −0.250

All crisis troughs show strong anti-correlation between COT and return clusters, but the economic
mechanism differs between acute crashes and gradual shock onsets.

#### Acute crash troughs — GFC 2008 (2008-01) and COVID 2020 (2020-02)

These show the clearest anti-prediction driven by forced liquidation. GFC 2008 trough is the most
extreme example — managed money were liquidating everything simultaneously, grouping CL, RB,
NG, HG together in their COT positioning while returns showed a completely different structure
driven by which assets could be sold fastest rather than fundamental views. COVID 2020 trough
(ARI −0.250, slightly different from −0.246) confirms this pattern as an independent observation
— February 2020 positioning reflected a pre-COVID world that became completely irrelevant within
weeks. During acute crashes, **COT positioning reflects liquidity constraints not fundamental views**
— the signal is not just absent, it is inverted.

#### Gradual shock onset troughs — Oil crash 2014 (2014-11) and Gas shock 2021 (2021-02)

These troughs have a different economic explanation. The anti-prediction here is not caused by
forced liquidation but by **regime transition timing**. In November 2014, the OPEC announcement
reconfigured commodity co-movement patterns faster than the 2-month lag could capture — the
September 2014 COT positioning reflected a pre-OPEC world that became misaligned with the new
return structure. In February 2021, managed money were still positioned in a post-COVID recovery
framework while early energy supply tensions were already reshaping return co-movement. The
managed money were not wrong about the direction — they were caught in the transition moment
where old positioning met a newly emerging return structure.

> **Key pattern:** crisis troughs produce ARI ≈ −0.246 through two distinct mechanisms —
> forced liquidation during acute crashes, and regime transition timing during gradual shock onsets.

## Calm Periods — Peak Signal Weeks
### Peak Calm — ARI 0.609 to 0.643

All three calm peaks (Baseline 2016, 2018, 2023) show consistent and economically coherent signal.
Contrary to what one might expect, the COT clusters during calm periods are just as cross-sectoral
as during gradual crisis peaks — metals mixed with agri, energy mixed with cereals. Baseline 2016
groups GC, HG, SI, NG together (precious metals + industrial metals + gas) and CL with ZW (crude
oil with wheat). Baseline 2018 groups SI, CL, ZW, GC, ZS on the same side — silver with crude oil,
wheat, gold and soybean. Baseline 2023 groups NG, ZW, CL together and HG, ZS, ZC together.
**None of these reflect natural sector boundaries.**

These cross-sectoral groupings are driven by quiet macro themes that cut across commodity
sectors — reflation and dollar weakness in 2016, energy-food inflation linkages in 2023,
dollar-driven precious metals positioning in 2018. Managed money build these cross-sectoral views
even in low-volatility environments, and the subsequent return co-movement confirms them with
moderate but consistent precision.

ARI values are slightly lower than gradual crisis peaks (0.609–0.643 vs 0.739) not because the
positioning is more sectoral, but because the macro themes are less dominant in calm periods —
idiosyncratic commodity factors matter more in low-volatility environments, introducing noise that
partially obscures the cross-sectoral signal.

> **Key pattern:** during calm regimes, managed money build cross-sectoral positioning driven by
> quiet macro themes — reflation, dollar moves, energy-food linkages — that predict return
> co-movement at moderate but consistent precision.

## Calm Periods — Trough Signal Weeks
### Trough Calm — ARI −0.246 (all)

All calm troughs converge to exactly ARI = −16/65 ≈ −0.246. This mathematical precision is not
a coincidence — it reflects a recurring structural configuration where each COT cluster splits exactly
across two return clusters. During calm periods, the weeks with the worst signal show a highly
repetitive and homogeneous anti-correlated pattern — managed money repositioning in low-volatility
weeks follows mechanical, predictable routines that systematically misalign with return co-movement
during those specific weeks.

Unlike crisis troughs where the anti-prediction reflects forced liquidation, calm troughs reflect a
different phenomenon — the COT signal in quiet weeks is **not noisy or chaotic, it is simply
structured in the wrong direction** in a consistent and mathematically precise way.

> **Key pattern:** calm trough weeks show a recurring anti-correlated structural configuration,
> distinct from crisis troughs in that it is homogeneous and repetitive rather than chaotic —
> suggesting a systematic misalignment between managed money routine positioning and return
> co-movement during specific low-activity weeks.

## Overall Conclusion Across All Four Regimes

The COT signal is strongest and most informational during **gradual commodity cycles** where
managed money build sustained cross-sectoral views driven by dominant macro themes — energy
supply shocks, commodity supercycles — producing peak ARI values up to 0.739. It is consistently
present but moderate during **calm periods** where managed money build quieter cross-sectoral
views around softer macro themes such as reflation, dollar moves, and energy-food linkages
(peak calm, ARI 0.609–0.643). In both cases the positioning is cross-sectoral rather than sectoral
— the difference lies in the intensity and dominance of the macro theme, not in the structural
logic of the positioning.

The signal inverts through two distinct mechanisms:

1. During **acute crashes** (GFC 2008, COVID 2020), forced liquidation drives COT positioning
to reflect liquidity constraints rather than fundamental views — the signal is not just absent, it is
inverted by panic selling (crisis troughs, ARI ≈ −0.246).

2. During **gradual shock onsets** (Oil crash 2014-11, Gas shock 2021-02), the 2-month lag
captures the exact transition moment where pre-shock COT positioning becomes structurally
misaligned with a newly reconfigured return co-movement regime — the managed money were
not wrong, but a structural break rendered their positioning outdated before it could materialize.

The **within-subperiod variance** of the signal is itself a key finding. The same subperiod contains
weeks with ARI up to 0.739 and weeks as low as −0.246 — a range of nearly one full unit. This
high variance suggests the signal is driven by episodic alignment events rather than persistent
regime-level predictability.

> The signal is therefore **regime-dependent, episodic, and only exploitable** when managed
> money positioning reflects genuine informational conviction about cross-commodity macro themes
> — rather than liquidity constraints, panic selling, structural break timing mismatches, or
> mechanical low-activity routines.

# Part II — Quantitative Results

---
## Signal Strength Across Horizons

Lift is defined as ARI_h − ARI_0. At h=1m the lift is negative (−0.002) — COT from 1 month ago
is slightly worse than the synchronous COT. At h=2m the lift is positive (+0.006) — COT from 2
months ago predicts return co-movement better than contemporaneous COT. At h=3m the lift
remains positive (+0.002) but weaker.

The sign change between h=1m and h=2m is the key finding. It suggests that the managed money
positioning signal takes approximately **2 months to materialize** in return co-movement —
consistent with typical managed money holding periods of 6–12 weeks. The h=1m horizon is too
short for the positioning to have generated sufficient price impact, while h=3m is starting to decay
as the positioning framework ages.

---
## Average Predictive Power

Mean ARI across all lags hovers between 0.041 and 0.049. The pattern is not monotonic — ARI
dips at h=1m (0.041) below the synchronous baseline (0.043), then peaks at h=2m (0.049) above
the baseline, then partially recovers at h=3m (0.045). The **inverted-U shape centered on h=2m**
is the first quantitative evidence that if a predictive signal exists, it operates specifically at a
2-month horizon rather than at shorter or longer horizons.

The values are small in absolute terms — all below 0.05 — but the shape of the curve matters as
much as the level. A flat curve would suggest no lag structure. The dip-peak-recovery pattern
suggests a genuine 2-month transmission mechanism.


image

---
## Regime-Dependent Dynamics

The aggregate null result (mean ARI ≈ 0.043–0.049) masks fundamentally different signal dynamics
across regimes. Three distinct patterns emerge:

**Pattern 1 — Genuine lead-lag signal during gradual commodity cycles**
Oil crash 2014-2015 and Gas shock 2021-2022 are the only subperiods where ARI increases with
lag. Oil crash shows ARI rising from 0.060 at h=2m to 0.073 at h=3m. Gas shock shows ARI rising
monotonically from 0.006 at h=0 to 0.062 at h=3m. This increasing-with-lag pattern is the clearest
empirical evidence that managed money positioning genuinely anticipates return co-movement
during sustained commodity cycles.

**Pattern 2 — Signal inversion during acute shocks**
GFC 2008-2009 and COVID 2020 show the opposite pattern — positive ARI at h=0 that turns
negative at longer lags. This confirms the forced liquidation interpretation — any COT positioning
from 2-3 months prior was invalidated by the subsequent market dislocation.

**Pattern 3 — Contemporaneous dominance during calm periods**
Baseline 2023 shows the highest ARI in the entire heatmap (0.093) but only at h=0 — collapsing
to near zero at longer lags. During calm periods the signal is real but faster-materialising —
consistent with a momentum interpretation where COT clusters reflect current institutional flows.

> **The single key conclusion:** the optimal lag is regime-dependent — 3 months during gradual
> commodity cycles, 1 month during calm periods, and zero or negative during acute shocks.
> There is no single universal lag at which the signal is consistently positive across all regimes.

## COT Category Decomposition

**M5_combined** (all categories concatenated) dominates at all lags — 0.073 at h=0, peaking at
0.075 at h=1m and remaining stable at 0.074–0.075 through h=3m. The combined specification is
both stronger and more stable across lags than any individual category.

**M1_MM** (Managed Money) peaks at h=2m (0.049) — confirming the 2-month optimal horizon
for speculative positioning.

**M2_PM** (Producer/Merchant) shows a relatively stable pattern with a slight peak at h=2m.
Commercial hedgers' positioning materialises at longer horizons than speculative positioning,
consistent with their role as fundamental hedgers who respond to physical supply-demand dynamics.

**M3_SD** (Swap Dealers) is the weakest category throughout — rising from 0.021 at h=0 to only
0.025 at h=1m then declining. Swap Dealer positioning adds minimal predictive information.

**M4_OR** (Other Reportables) shows a declining pattern from 0.034 at h=0 to 0.028 at h=3m —
positioning reflects reactive rather than anticipatory behavior.

> **Critical insight:** combining all categories (M5_combined) substantially outperforms any
> individual category, confirming that predictive information about cross-commodity co-movement
> is distributed across all trader types rather than concentrated in speculative flows alone.

## Significance Testing

All four lags show statistically significant positive mean ARI after Newey-West HAC correction for
serial autocorrelation (t-statistics ranging from 5.32 to 5.79, all p < 0.001). The confidence
intervals are wide — reflecting the high episodic variance of the weekly ARI series — but their
lower bounds remain comfortably above zero at every lag.

The t-statistic peaks at h=2m (t = 5.79) — the highest of all four lags — confirming that the
2-month horizon is not only the strongest in mean ARI (0.049) but also the most precisely
estimated and statistically robust predictive window.

*Note on HAC correction:* The 52-week rolling window used to build COT features introduces
strong serial autocorrelation in the ARI series. Standard OLS standard errors would severely
understate uncertainty. Newey-West HAC standard errors are consistent under arbitrary
heteroskedasticity and autocorrelation, with bandwidth automatically selected as
nlags = ⌊4 × (T/100)^(2/9)⌋ ≈ 6 lags.

> **Key pattern:** COT positioning clusters statistically significantly predict return co-movement
> clusters at all tested horizons from synchronous to 3 months, with the **2-month lag representing
> the optimal and most robust predictive window** (t = 5.79, p < 0.001).

## Conclusion

COT positioning clusters **statistically significantly predict** commodity return co-movement
clusters at a 2-month horizon (t = 5.79, p < 0.001, Newey-West HAC corrected), with mean ARI
peaking at 0.049 at h=2m.

The signal is **regime-dependent and episodic**:
- **Strongest** during gradual commodity cycles where managed money build sustained
  cross-sectoral views over several months (ARI up to 0.739 at peak weeks)
- **Absent or inverted** during acute crashes where forced liquidation dominates
- **Faster-materialising** during calm periods where quieter macro themes transmit in
  approximately 1 month rather than 2

The COT category decomposition reveals that **M5_combined** (all four trader categories
concatenated) substantially outperforms M1_MM alone across all lags (ARI 0.073–0.075 vs
0.041–0.049), confirming that predictive information about cross-commodity co-movement is
distributed across the full institutional positioning landscape.

The **within-subperiod variance** is itself a key structural finding. Peak weeks produce ARI up to
0.739, while trough weeks produce ARI as low as −0.246. The signal exists when the macro
framework embedded in COT positioning at t−2m survives intact to t, and inverts when any
disruption intervenes within that window.

The consistently positive **synchronous ARI (h=0)** across all subperiods — including crisis
periods where the lagged signal breaks down — raises an intriguing open question: COT
positioning may be better understood as a **momentum indicator** reflecting active institutional
flows currently driving commodity co-movement, rather than as a predictor of future structure.
Whether this momentum interpretation translates into a viable trading signal remains an open
empirical question for future work.
